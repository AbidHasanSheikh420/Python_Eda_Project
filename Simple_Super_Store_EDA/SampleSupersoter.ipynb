```python
#Missing Value Analysis 
data.isnull().sum() 
```

```python
sns.heatmap(data.isnull(), cbar=True) 
plt.show() 
```

```python
#Duplicate Value Check 
data.duplicated().sum() 
data.drop_duplicates(inplace = True) 
```

```python
#Univariate Analysis (One Column) 
sns.histplot(data["Sales"], kde=True)
plt.show() 
```

```python
#Categorical wise count 
sns.countplot(x="Category", data=data) 
plt.show() 
```

```python
# Bivariate Analysis 
sns.barplot(x="Category", y="Sales", data=data) 
plt.show( )
```

```python
sns.boxplot(x="Category", y="Profit", data=data) 
plt.show() 
```

```python
sns.scatterplot(x="Sales", y="Profit",  data = data) 
plt.show( )
```

```python
corr = data.corr(numeric_only = True) 
sns.heatmap(corr, annot=True, cmap="coolwarm") 
plt.show() 
```

```python
#Time  Serise Analysis (IF data Column Existt) 
data["Order Date"] = pd.to_datetime(data["Order Date"])

data.groupby("Order Date")["Sales"].sum().plot() 
plt.show() 
```
