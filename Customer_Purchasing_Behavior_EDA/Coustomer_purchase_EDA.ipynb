```python
msno.matrix(data)
msno.bar(data) 
msno.heatmap(data) 
```

```python
#Age Distribution 
sns.histplot(data["age"], kde=True)
plt.title("Age Distribution") 
plt.show() 
```

```python
sns.histplot(data["annual_income"], bins=30)
plt.title("Annual INcome Distribution") 
plt.show()
```

```python
sns.countplot(x="purchase_frequency", data=data)
plt.title("Purchase Frequency Distribution") 
plt.show()
```

```python
data.columns
```

```python
data["purchase_amount"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Payment Method Share") 
plt.ylabel("") 
plt.show() 
```

```python
sns.countplot(x="loyalty_score", data=data)
plt.title("Customer Loyalty Status") 
plt.show()
```

```python
sns.boxplot(x="loyalty_score", y="purchase_amount", data=data)
plt.title("Loyalty vs Purchase Amount")
plt.show() 
```
