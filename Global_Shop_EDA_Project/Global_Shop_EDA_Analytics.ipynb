```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt



df = pd.read_csv("global_superstore_2016.csv")

```

```python
df.head()
df.shape
df.info()
df.describe()
df.isnull().sum()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Order Month Name'] = df['Order Date'].dt.month_name()

df['Shipping Days'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
sns.histplot(df['Sales'], bins=30, kde=True)
plt.title("Sales Distribution")
plt.show()
```

```python
sns.histplot(df['Profit'], bins=30, kde=True)
plt.title("Profit Distribution")
plt.show()
```

```python
sns.countplot(x='Quantity', data=df)
plt.title("Quantity Distribution")
plt.show()
```

```python
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title("Sales vs Profit")
plt.show()
```

```python
sns.barplot(x='Category', y='Sales', data=df, estimator=np.sum)
plt.title("Category-wise Sales")
plt.show()
```

```python
sns.barplot(x='Category', y='Profit', data=df, estimator=np.sum)
plt.title("Category-wise Profit")
plt.show()
```

```python
sns.countplot(x='Ship Mode', data=df)
plt.title("Shipping Mode Distribution")
plt.show()
```

```python
sns.histplot(df['Shipping Cost'], bins=30, kde=True)
plt.title("Shipping Cost Distribution")
plt.show()
```

```python
sns.countplot(x='Order Priority', data=df)
plt.title("Order Priority Distribution")
plt.show()
```
