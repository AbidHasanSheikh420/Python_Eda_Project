```python
#Revenue by product type 

rev_product = data.groupby("Product type")["Revenue generated"].sum() 
rev_product.plot(kind= "bar") 
plt.title("Revenue by product Type") 
plt.xlabel("Product Type") 
plt.ylabel("Revenue") 
plt.show() 
```

```python
#Sales performance 

sales_location = data.groupby("Location")["Number of products sold"].sum() 
sales_location.plot(kind= "bar")
plt.title("Product Sold by Location") 
plt.show() 
```

```python
#Inventory / Stock Analysis

stock_sku = data.groupby("SKU")["Stock levels"].mean() 
stock_sku.plot(kind= "line") 
plt.title("Average Stock Level by SKU") 
plt.show() 
```

```python
#Manufacturing Cost Analysis
manu_cost = data.groupby("Supplier name")["Manufacturing costs"].sum() 

manu_cost.plot(kind= 'bar') 
plt.title("Manufacturing Cost by Supplier") 
plt.show() 
```

```python
#Manufacturing lead time analysis 
plt.hist(data["Manufacturing lead time"], bins=10) 
plt.title("Manufacturing Lead Time Distribution") 
plt.xlabel("Days") 
plt.ylabel("Frequency") 
plt.show() 
```

```python
#Shipping & Logistics
ship_cost = data.groupby("Shipping carriers")["Shipping costs"].sum() 
ship_cost.plot(kind= "bar") 
plt.title("Shipping Cost by Carrier") 
plt.show() 
```

```python
#Defect & Quality Analysis 
plt.boxplot(data["Defect rates"])
plt.title("Defect Rate Distribution'") 
plt.show() 
```
